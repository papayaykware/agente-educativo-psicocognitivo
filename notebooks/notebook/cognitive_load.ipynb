# Estimación de Carga Cognitiva
Autor conceptual: Copilot

Este notebook permite:
- Calcular carga cognitiva heurística
- Visualizar evolución temporal
- Integrar métricas con el perfil psicocognitivo

```
def cognitive_load(errors, response_time, lexical_complexity):
    return 0.4*errors + 0.3*response_time + 0.3*lexical_complexity

cognitive_load(0.2, 0.5, 0.3)
```

```
import matplotlib.pyplot as plt

loads = [cognitive_load(0.1*i, 0.05*i, 0.02*i) for i in range(10)]
plt.plot(loads)
plt.title("Evolución de carga cognitiva")
plt.show()
```
